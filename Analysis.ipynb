```python
from __future__ import unicode_literals

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rcParams
%matplotlib inline
rcParams['figure.figsize'] = (100, 10)
rcParams['font.family'] = 'AppleGothic'
```

```python
df12 = pd.read_csv('12-KAPF-1_utf.csv', skiprows=1, thousands=',', names=['TID', 'MemberID', 'Name', 'Party', 'Area', 'Transaction Date', 'Transaction Detail', 'Transaction Amount', 'Target', 'Category'])
df13 = pd.read_csv('13-KAPF-1_utf.csv', skiprows=1, thousands=',', names=['TID', 'MemberID', 'Name', 'Party', 'Area', 'Transaction Date', 'Transaction Detail', 'Transaction Amount', 'Target', 'Category'])
df14 = pd.read_csv('14-KAPF-1_utf.csv', skiprows=1, thousands=',', names=['TID', 'MemberID', 'Name', 'Party', 'Area', 'Transaction Date', 'Transaction Detail', 'Transaction Amount', 'Target', 'Category'])
frames = [df12, df13, df14]
df = pd.concat(frames)
df
```

```python
df["Transaction Amount"]
```

```python
df["Transaction Amount"] = pd.to_numeric(df["Transaction Amount"])
```

```python
members = df[["Name", "Transaction Amount"]].groupby("Name")
member_totals = members.sum()
```

```python
member_totals.sort_values(by='Transaction Amount')
```

```python
member_totals.sort_values(by='Transaction Amount', ascending=False).plot(kind="bar")
```

```python
categories = df[["Category", "Transaction Amount"]].groupby("Category")
category_totals = categories.sum()
```

```python
category_totals.sort_values(by='Transaction Amount', ascending=False).plot(kind="bar")
```
